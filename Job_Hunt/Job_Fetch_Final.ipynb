```python
from bs4 import BeautifulSoup
from splinter import Browser
import time
import requests
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

The following loop goes to different job search pages by city and then grabs 2 pages worth of jobs after selecting the different job categories. Note sf and seattle had different structured pages so they didn't pull any results.

```python
#Create lists to hold data to put into columns for a dataframe
Job_Title = []
Company = []
City = []
Field = []
Tech_Hub = []
#loop of Tech Hubs, but changed to 5 because the hubs of sf and seattle don't have the same page structure.
city_loop = ["chicago","boston","austin","colorado","nyc"]
#List of job fields that I wish to click through as I loop through the different Tech Hubs
Fields = ["Data + Analytics","Design","Dev + Engineer","Marketing","Project Mgmt","Product"]
#loop to go through the different tech hubs
for c in city_loop:
    #logic adapt urls based on the tech hub desired. logic to handle the fact that chicago was .org and the rest were .com
    if c in ["chicago"]:
        url =f"https://www.builtin{c}.org/jobs"
    else: 
        url =f"https://www.builtin{c}.com/jobs"
    # browse to the tech hub url
    browser.visit(url)
    # slight delay to let the page load
    time.sleep(0.25)
    #loop to go through and click the different job field options
    for field in Fields:
        #print field <-- I had printed field to track progress and know if it got hung up somewhere. 
        #How I found out sf and seattle didn't function
        #click field buttons
        browser.links.find_by_partial_text(field)
        #pause to let the page load
        time.sleep(2)
        
        #ultimate range just allows me to tell how many pages I would like to go through because at the end of the loop 
        #the page moves to the next page.  
        for i in range(2):
            
            #read and store html as a soup object
            soup = BeautifulSoup(html, 'html.parser')
            #grab all subsections of "left-col" which are the containers with the job postings and store in a list
            comp_jobs = soup.find_all('div', class_="left-col")
            
            #Loop through the subsections and do so as integers so that list items can be referenced
            for k in range(len(comp_jobs)):
                
                #store title, company, and city
                title = comp_jobs[k].find('h2',class_="title").text
                company=comp_jobs[k].find('div',class_="company-title").text
                city=comp_jobs[k].find('div',class_="job-location").text
                #I stored them and printed them to know progress as it was running because it took about 10 minutes to run.
#                 print(title)
#                 print(company)
#                 print(city)
#                 print(c)
#                 print("----------------------------")
                #appended lists with each jobs information and at the same time stored current field and tech hub
                Field.append(field)
                Job_Title.append(title)
                Company.append(company)
                City.append(city)
                Tech_Hub.append(c)
            #click next button to go to next job page for additional scraping
            browser.links.find_by_partial_text('Next')
            #pause for page to load
            time.sleep(2)
    
```

This code visits the seattle jobs page because it was formatted differently than the the other pages.  It loops through the different categories and for each category grabs job information of 2 pages.  The loop logic is the same as the others after you get to the url, but the scrape references are different.

```python
url ="https://www.builtinseattle.com/jobs"
browser.visit(url)
#     print(c)
time.sleep(0.25)

for field in Fields:
        browser.links.find_by_partial_text(field)
        time.sleep(2)
        print(field)
        for i in range(2):
            html = browser.html
            #read and store html
            soup = BeautifulSoup(html, 'html.parser')
            comp_jobs = soup.find_all('div', class_="condensed-job-row")
            print(comp_jobs)    
            for k in range(len(comp_jobs)):
                
                title = comp_jobs[k].find('h2', class_="job-title").text
                
                company=comp_jobs[k].find_all('span',class_="param-value")[0].text.strip()
                city=comp_jobs[k].find_all('span',class_="param-value")[1].text.strip()
                print(title)
                print(company)
                print(city)
                print(c)
                print("----------------------------")
                
                Field.append(field)
                Job_Title.append(title)
                Company.append(company)
                City.append(city)
                Tech_Hub.append("Seattle")
            browser.links.find_by_partial_text('Next')
            time.sleep(3)
```
