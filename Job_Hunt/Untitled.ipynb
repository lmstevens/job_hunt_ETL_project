```python
from bs4 import BeautifulSoup
from splinter import Browser
import time
import requests
import pandas as pd
```

Windows User

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://builtin.com/jobs"
browser.visit(url)
time.sleep(0.25)
Job_Title = []
Company = []
City = []
Field = []

Fields = ["Data + Analytics","Design","Dev + Engineer","Marketing","Project Mgmt","Product"]

for field in Fields:
    browser.links.find_by_partial_text(field)
    time.sleep(2)

    for i in range(10):
        html = browser.html
        #read and store html
        soup = BeautifulSoup(html, 'html.parser')
        job_chunk = soup.find_all('div', class_="job-item")

        for j in range(len(job_chunk)):
            comp_jobs = job_chunk[j].find_all('div',class_='condensed-job-row')

            for k in range(len(comp_jobs)):

                title = comp_jobs[k].find('h2',class_="job-title").text
                company=comp_jobs[k].find_all('span', class_="param-value")[0].text.strip()
                city=comp_jobs[k].find_all('span', class_="param-value")[1].text.strip()
                print(title)
                print(company)
                print(city)
                print("----------------------------")
                Field.append(field)
                Job_Title.append(title)
                Company.append(company)
                City.append(city)

        browser.links.find_by_partial_text('Next')
        time.sleep(3)
```

```python
print(len(City))
print(len(Company))
print(len(Job_Title))
print(len(Field))
```

```python
df= pd.DataFrame({"Field":Field,"Job_Title":Job_Title,"City":City,"Company":Company})
df.groupby([Field,City]).count()
```

```python
url = "https://builtin.com/tech-hubs"
browser.visit(url)
time.sleep(0.25)
Job_Title = []
Company = []
City = []
Field = []
Tech_Hub = []
city_loop = ["chicago","boston","austin","colorado","nyc","sf","seattle"]
Fields = ["Data + Analytics","Design","Dev + Engineer","Marketing","Project Mgmt","Product"]
for c in city_loop:
    if c in ["chicago"]:
        url =f"https://www.builtin{c}.org/jobs"
    else: 
        url =f"https://www.builtin{c}.com/jobs"
    print(url)
    browser.visit(url)
    print(c)
    time.sleep(0.25)
    for field in Fields:
        browser.links.find_by_partial_text(field)
        time.sleep(2)
        print(field)
        for i in range(2):
            html = browser.html
            #read and store html
            soup = BeautifulSoup(html, 'html.parser')
            comp_jobs = soup.find_all('div', class_="left-col")
            print(i)    
            for k in range(len(comp_jobs)):
                
                title = comp_jobs[k].find('h2',class_="title").text
                company=comp_jobs[k].find('div',class_="company-title").text
                city=comp_jobs[k].find('div',class_="job-location").text
                print(title)
                print(company)
                print(city)
                print(c)
                print("----------------------------")
                
                Field.append(field)
                Job_Title.append(title)
                Company.append(company)
                City.append(city)
                Tech_Hub.append(c)
            browser.links.find_by_partial_text('Next')
            time.sleep(3)
    
```
