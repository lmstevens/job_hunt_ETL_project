```python
from bs4 import BeautifulSoup
from splinter import Browser
import time
import requests
import os
from selenium import webdriver
import pandas as pd
```

```python
url = "https://builtin.com/tech-hubs"
browser = webdriver.Chrome()
browser.get(url)
time.sleep(10)
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
#click link to tech hubs selenium full xpath code
#browser.find_element_by_xpath("""/html/body/div[1]/div/header/div/div/div/div/div[1]/div[1]/div/div[2]/div/div/div[1]/div[3]/a""").click()
```

```python
#find and collect all cities listed on page
cities = browser.find_elements_by_xpath('//div[@class="block block-techhub-cost-block"]//div[@class="block-content"]//div[@class="widget"]//div[@class="title"]')

cities2 = []

for i in range(len(cities)):
    cities2.append(cities[i].text)

del cities2[1::2]
print(cities2)
```

```python
#find and collect all cities and their cost of living for 1 bedroom 

living_cost = browser.find_elements_by_xpath('//div[@class="block block-techhub-cost-block"]//div[@class="block-content"]//div[@class="widget"]//div[@class="living-data"]//div[@class="living rent"]//div[@class="value"]')

living_cost2 = []

for i in range(len(living_cost)):
    living_cost2.append(living_cost[i].text)

del living_cost2[1::2]
print(living_cost2)
```

```python
#find and collect all cities and their cost of avg lunch

lunch_cost = browser.find_elements_by_xpath('//div[@class="block block-techhub-cost-block"]//div[@class="block-content"]//div[@class="widget"]//div[@class="living-data"]//div[@class="living lunch"]//div[@class="value"]')

lunch_cost2 = []

for i in range(len(lunch_cost)):
    lunch_cost2.append(lunch_cost[i].text)

del lunch_cost2[1::2]
print(lunch_cost2)
```

```python
#find and collect all cities and their median home price

home_cost = browser.find_elements_by_xpath('//div[@class="block block-techhub-cost-block"]//div[@class="block-content"]//div[@class="widget"]//div[@class="living-data"]//div[@class="living home-price"]//div[@class="value"]')

home_cost2 = []

for i in range(len(home_cost)):
    home_cost2.append(home_cost[i].text)

del home_cost2[1::2]
print(home_cost2)
```

```python
#find and collect all cities populations

population = browser.find_elements_by_xpath('//div[@class="block block-techhub-all-block"]//div[@class="views-field views-field-population"]//div[@class="field-content"]')

pop2 = []

for i in range(len(population)):
    pop2.append(population[i].text)

print(pop2)
```

```python
#number of tech companies per city

tech_comp = browser.find_elements_by_xpath('//div[@class="block block-techhub-all-block"]//div[@class="views-field views-field-companies-number"]//div[@class="field-content"]')

tech_comp2 = []

for i in range(len(tech_comp)):
    tech_comp2.append(tech_comp[i].text)

print(tech_comp2)
```

```python
#most popular fields of tech in each city

popular_fields = browser.find_elements_by_xpath('//div[@class="views-field views-field-field-th-industries"]//div[@class="paragraph paragraph--type--tech-hub-industry paragraph--view-mode--default"]//div[@class="field--name-field-th-industry-name field--type-string field__item"]')       

popular_fields2 = []

for i in range(len(popular_fields)):
    popular_fields2.append(popular_fields[i].text)

print(popular_fields2)
```

```python
pop_field = pd.DataFrame({"popular_fields2":popular_fields2})
pop_field2 = pop_field.popular_fields2.unique()
#len(popular_fields2)
pop_field2
```
